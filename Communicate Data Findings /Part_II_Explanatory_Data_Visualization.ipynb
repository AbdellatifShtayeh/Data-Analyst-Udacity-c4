## Investigation Overview
This presentation aims to highlight key trends and relationships in the Prosper Loan Data, focusing on loan performance, borrower rates, and borrower characteristics. The goal is to provide clear, actionable insights for stakeholders.

**Executive Summary:**
- Higher borrower rates are associated with increased default rates.
- Prosper rating is a strong predictor of loan outcomes.
- Borrower income influences loan amounts and rates.

## Dataset Overview and Executive Summary
The Prosper Loan Data dataset includes 113,937 loans with 81 variables, such as loan amount, borrower rate, loan status, and borrower income. This analysis distills the data into key findings:

1. Borrower rates peak around 0.15-0.20.
2. Higher rates correlate with defaults.
3. Prosper rating effectively indicates loan risk.

```python
# Import packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Suppress warnings
import warnings
warnings.simplefilter('ignore')

# Load the dataset
df = pd.read_csv('Data/prosperLoanData.csv')
```

```python
# Visualization 1: Histogram of Borrower Rate
plt.figure(figsize=(10, 6))
sns.histplot(df['BorrowerRate'], bins=30, kde=True, color='blue')
plt.title('Distribution of Borrower Rates')
plt.xlabel('Borrower Rate')
plt.ylabel('Frequency')
plt.show()
```

**Comment:** The borrower rate distribution is roughly normal, with most loans clustering between 0.15 and 0.20, indicating a common interest rate range.

```python
# Visualization 2: Box Plot of Borrower Rate by Loan Status
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='LoanStatus', y='BorrowerRate', palette='Set2')
plt.title('Borrower Rate by Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Borrower Rate')
plt.xticks(rotation=45)
plt.show()
```

**Comment:** Defaulted and charged-off loans exhibit higher median borrower rates, suggesting that riskier loans command higher interest.

```python
# Visualization 3: Clustered Bar Chart of Prosper Rating and Loan Status
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ProsperRating (Alpha)', hue='LoanStatus', order=['AA', 'A', 'B', 'C', 'D', 'E', 'HR'], palette='viridis')
plt.title('Loan Status by Prosper Rating')
plt.xlabel('Prosper Rating')
plt.ylabel('Count')
plt.legend(title='Loan Status')
plt.show()
```

**Comment:** Higher Prosper ratings (AA, A) show fewer defaults, while lower ratings (E, HR) have more, demonstrating the rating system's reliability in assessing risk.
